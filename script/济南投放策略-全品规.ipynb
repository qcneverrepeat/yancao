```python
import numpy as np
import pandas as pd
```

# 读入整理后的数据

```python
data = pd.read_csv('jinan+.csv')
data.head()
```

```python
data1 = data.copy()
```

删去带‘-’的无效数据，目前认为对这些档位不做调整

```python
data_clean = data1.replace('-', 0, inplace=False)
```

```python
data_clean['政策变动'] = data_clean['下期政策'] - data_clean['本期政策']
```

```python
data_clean.head()
```

构建机器学习模型，自变量为'档位','需求满足率','订购率','订足面','订足率(济南)','户均条数','价位'，因变量为'政策变动'

```python
df = data_clean[['本期政策','档位','需求满足率','订购率','订足面','订足率（中烟）','订足率(济南)','户均条数','价位','是否省内','政策变动']]
df.head()
```

# 数据分析模块（决策树）

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction import DictVectorizer #特征转换器
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn import tree
```

```python
X = df[['本期政策','档位','需求满足率','订购率','订足面','订足率（中烟）','订足率(济南)','户均条数','价位','是否省内']]
y = df['政策变动']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=2)  # 将数据进行分割
```

```python
# vec = DictVectorizer(sparse=False)
# X_train = vec.fit_transform(X_train.to_dict(orient='record'))   #对训练数据的特征进行提取
# X_test = vec.transform(X_test.to_dict(orient='record'))         #对测试数据的特征进行提取
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_train,y_train)
y_predict = dtc.predict(X_test)
```

```python
print('Accracy:',dtc.score(X_test,y_test))
print(classification_report(y_predict,y_test))
```
