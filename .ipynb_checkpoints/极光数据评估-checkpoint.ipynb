```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 极光匹配群体 vs 联通匹配群体

```python
# 965个手机号中选出587个
jg = pd.read_excel('./data/标签匹配&川大.xlsx',sheet_name='详情')
jg = jg[jg['查询结果']=='查询标签成功-有标签']

# 取部分有用的字段
sel = ['imei/phone/mac/imsi','CPL_INDM_GEND_S','CPL_INDM_AGE_C5','SOM_OCM_CAREER','CPL_INDM_EDU_LEVEL','CPL_HHM_CHILD_HC','CPL_INDM_VEIC_VEID','FIM_FISM_INCL','FIM_FISM_CONL_CIR','GBM_BHM_APPP_APPR_S','GBM_BHM_PURB_PURW'] # 选择和联通平台对比的字段
re = jg.loc[:,sel]

print(jg.shape, re.shape)
```

```python
# 1.性别比例
na = 587 - re['CPL_INDM_GEND_S'].isna().sum()
gender = re['CPL_INDM_GEND_S'].dropna().value_counts()/re['CPL_INDM_GEND_S'].dropna().shape[0]

print('极光平台数据有效条数: %d'%na)
print('极光平台得到的性别比例：男 %.2f，女 %.2f'%(gender['M'],gender['F']))
print('联通平台得到的性别比例：男 0.88，女 0.12')
```

```python
# 2.年龄分布: 两平台粒度不一致
na = 587 - re['CPL_INDM_AGE_C5'].isna().sum()
age = re['CPL_INDM_AGE_C5'].dropna().value_counts()/re['CPL_INDM_AGE_C5'].dropna().shape[0]

print('极光平台数据有效条数: %d'%na)
print('极光平台得到的年龄分布：0~15岁  %.2f, 16~25岁 %.2f, 36~45岁 %.2f, ⼤于46岁 %.2f'%(age[1.0],age[2.0],age[3.0],age[4.0]))
print('联通平台得到的年龄分布：19~29岁 0.08, 30~39岁 0.31, 40~49岁 0.33, 50~59岁 0.21, 60岁以上 0.07')
```

```python
# 3.(京东)子女情况: 两平台字段取值不一致
na = 587 - re['CPL_HHM_CHILD_HC'].isna().sum()
ch = re['CPL_HHM_CHILD_HC'].dropna().value_counts()/re['CPL_HHM_CHILD_HC'].dropna().shape[0]

print('极光平台数据有效条数: %d'%na)
print('极光平台得到的是否有子女: 有 %.2f, 无 %.2f'%(ch['有'], ch['无']))
print('联通平台得到的是否有子女: 高可能性 0.33, 中可能性 0.45, 低可能性 0.22')
```

```python
# 4.(京东)职业: 极光平台数据量过少,两平台字段取值不一致
na = 587 - re['SOM_OCM_CAREER'].isna().sum()
job = re['SOM_OCM_CAREER'].dropna().value_counts()/re['SOM_OCM_CAREER'].dropna().shape[0]

print('极光平台数据有效条数: %d，且两平台字段取值差异太大，未比较'%na)
```

```python
# 5.(京东)学历: 两平台字段取值不一致
na = 587 - re['CPL_INDM_EDU_LEVEL'].isna().sum()
edu = re['CPL_INDM_EDU_LEVEL'].dropna().value_counts()/re['CPL_INDM_EDU_LEVEL'].dropna().shape[0]

print('极光平台数据有效条数: %d'%na)
print('极光平台得到的学历情况: 低 %.2f, 中 %.2f, 高 %.2f'%(edu['低'], edu['中'], edu['高']))
print('联通平台得到的学历情况: 0.46 为高中及以下, 0.54 拥有本科学历')
```

```python
# 6.(京东)综合消费能力
na = 587 - re['FIM_FISM_CONL_CIR'].isna().sum()
cos = re['FIM_FISM_CONL_CIR'].dropna().value_counts()/re['FIM_FISM_CONL_CIR'].dropna().shape[0]

print('极光平台数据有效条数: %d'%na)
print('极光平台得到的综合消费能力分布: 高综合消费能力 %.2f, 中等综合消费能力 %.2f, 低综合消费能力 %.2f'%(cos['高'],cos['中'],cos['低']))
print('联通平台得到的综合消费能力分布: 高综合消费能力 0.76, 中等综合消费能力 0.21, 属于低综合消费能力 0.03')
```

```python
# 7.(京东)是否有车: 极光平台数据量过少,两平台字段取值不一致

na = 587 - re['CPL_INDM_VEIC_VEID'].isna().sum()
car = re['CPL_INDM_VEIC_VEID'].dropna().value_counts()/re['CPL_INDM_VEIC_VEID'].dropna().shape[0]
# 车辆预测中72%有车概率高，28%有车概率低。

print('极光平台数据有效条数: %d'%na)
print('极光平台得到的有车情况: 有车 %.2f'%(car['Y']))
print('联通平台得到的有车情况: 有车概率高 0.72, 有车概率低 0.28')
```

## 极光匹配群体（中的联通用户）vs 联通匹配群体

- 极光在整体会员手机号中匹配率587/965 = 61%，高于联通的221/965 = 23%

- 但极光在联通用户群中匹配率150/233 = 64% 要低于联通在联通用户群中221/233 = 95%的匹配率

- 所以此对比实为 这150名用户的画像(极光) 和 221名用户的画像(联通) 对比，仅供参考

- 标注了这150名用户中极光各字段的有效观测条数

- 选择的字段中,只有1.性别和2.年龄是联通自有的,其余均是京东侧数据

```python
jg = pd.read_excel('./data/标签匹配&川大.xlsx',sheet_name='详情')

# 965个手机号中选出联通用户：233个
lt = jg[jg['imei/phone/mac/imsi'].map(lambda x: str(x)[:3] in ['130','131','132','152','155','156','185','186'])]

# 233个联通用户中匹配到150个
sel = ['imei/phone/mac/imsi','CPL_INDM_GEND_S','CPL_INDM_AGE_C5','SOM_OCM_CAREER','CPL_INDM_EDU_LEVEL','CPL_HHM_CHILD_HC','CPL_INDM_VEIC_VEID','FIM_FISM_INCL','FIM_FISM_CONL_CIR','GBM_BHM_APPP_APPR_S','GBM_BHM_PURB_PURW'] # 选择和联通平台对比的字段
re = lt[lt['查询结果']=='查询标签成功-有标签'].loc[:,sel]

print(jg.shape, lt.shape, re.shape)
```

```python
# 1.性别比例
na = 150 - re['CPL_INDM_GEND_S'].isna().sum()
gender = re['CPL_INDM_GEND_S'].dropna().value_counts()/re['CPL_INDM_GEND_S'].dropna().shape[0]

print('极光平台数据有效条数: %d'%na)
print('极光平台得到的性别比例：男 %.2f，女 %.2f'%(gender['M'],gender['F']))
print('联通平台得到的性别比例：男 0.88，女 0.12')
```
